# Merge

```python
# for merging there should be something common,otherwise there will be no merging done
leftdf=pd.DataFrame({'alphabet':['a','b','c','d'],
                    'number':['1','2','3','4'],
                    'alphanumeric':['a1','a2','a3','a4']})
rightdf=pd.DataFrame({'alphabet':['a','b','c','d'],
                    'number':['5','6','7','8'],
                    'alphanumeric':['a5','a6','a7','a8']})
a=pd.merge(leftdf,rightdf,on='alphabet')
a
```

# join

```python
left=pd.DataFrame({'alphabet':['a','b','c','d'],
                    'number':['1','2','3','4'],
                    'alphanumeric':['a1','a2','a3','a4']})
right=pd.DataFrame({'alphabet':['a','b'],
                    'number':['5','6'],
                    'alphanumeric':['a5','a6']})
a=right.join(left,lsuffix='fdf',rsuffix='rdf')
a
```

# concatenate

```python
fdf=pd.DataFrame({'name':["A","b","c"],
                 'age':[15,16,27]})
sdf=pd.DataFrame({'name':["A","b","c"],
                 'age':[15,16,27]})
pd.concat([fdf,sdf])
```

```python
fdf1=pd.DataFrame({'name':["A","b","c"],
                 'age':[15,16,27]})
sdf1=pd.DataFrame({'naame':["A","b","c"],
                 'age':[15,16,27]})
pd.concat([fdf1,sdf1])
```
