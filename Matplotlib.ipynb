# Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np

#creating the data
a=np.linspace(0,10,20)
b=np.linspace(10,0,20)
#figure size
plt.figure(figsize=(4,4)) #4,4 is a ratio

#creating subplot
ax1=plt.subplot(121) #1-row 2-column 1-position
ax1.plot(a,linewidth=4,linestyle='-',color='r',marker='^')
ax2=plt.subplot(122)
ax2.plot(b)
plt.show()
```

```python
c=np.linspace(0,10,20)
plt.figure(figsize=(6,4))
ax3=plt.subplot()
plt.xlabel('No of points')
plt.ylabel('Data Values')
plt.title('lineplot')
plt.grid(True)

ax3.plot(a,linewidth=5,linestyle='-',color='r')
plt.savefig('basic.png')
plt.show()
```

```python
# draw a line plot in which there are five segments
a=np.linspace(0,5,20)
b=np.linspace(1,6,20)
c=np.linspace(12,34,20)
d=np.linspace(13,61,20)
e=np.linspace(14,62,20)
ax1=plt.subplot()
ax2=plt.subplot()
ax3=plt.subplot()
ax4=plt.subplot()
ax5=plt.subplot()
ax1.plot(a)
ax2.plot(b)
ax3.plot(c)
ax4.plot(d)
ax5.plot(e)
```

# scatter plot

```python
a=[10,20,30,40,50,60,70,80]
b=[1,2,3,4,3,2,1,2]
c=[15,60,75,80,44,35,55,77]

plt.figure(figsize=(4,4))
ax1=plt.subplot()
plt.scatter(b,a,s=400,edgecolor='r',color='g',marker='^',alpha=1)
plt.scatter(b,c,s=400,edgecolor='y',c='r',marker='o',alpha=1)
plt.grid()
plt.show()
```

# Box Plot

```python
# we can also use boxplot for finding the outliers
total=[20,4,1,30,20,10,20,70,30,10]
orders=[10,3,1,15,17,2,30,44,2,1]
discount=[30,20,10,5,20,10,60,20,40,1]
data=list([total,orders,discount])
print(data)
plt.figure(figsize=(4,4))
ax1=plt.subplot()
ax1.boxplot(data)
plt.show()

```

```python
# first line  from above stands for max value
#second line stands for third quaartile
#third-second quartile or median
#fourth-first quartile
#fifth-min value
```

# violin plot

```python
total=[20,4,1,30,20,10,20,70,30,10]
orders=[10,3,1,15,17,2,30,44,2,1]
discount=[30,20,10,5,20,10,60,20,40,1]
data=list([total,orders,discount])
plt.figure(figsize=(4,4))
ax1=plt.subplot()
ax1.violinplot(data)
plt.show()

# higher the thickness of the violin the more is the data
#lesser the thickness of the violin the less is the data ,the outlier will be there
```

# Histogram


```python
#basically histogram is used to present the frequency of the data
import matplotlib.pyplot as plt
a=[1,2,3,5,4,34,45,53,5,25,23,36,46,47,54,33,13,25]
plt.figure(figsize=(5,5))
ax1=plt.subplot()
ax1.hist(a)
plt.show() #by default we get only 10 bins or 10 bars
```

# pie chart

```python
# we are using the pie chart for percentage representation

lables=['frogs','crocodiles','bees','tigers']
sizes=[35,40,20,50]
explode=[0,0,0,0.1]

plt.figure(figsize=(5,5))
ax1=plt.subplot()
ax1.pie(sizes,labels=lables,explode=explode,autopct="%1.2f") #explode-cutting the piece or highliting the piece
plt.show()
```
