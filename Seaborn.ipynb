# seaborn

-built on matlplotlib
-fewer syntax
-we are unable to customize the plot in seaborn
-they will be using default themes
-basically used for statistical plotting
-using seaborn we will be getting interactive plotting


It is used for making statistical graphs using python
It is built on top of the matplotlib and is also compatible with pandas datas structure
limited scope of changing the things in seaborn
a single line code can work easily for plotting the graphs



```python
import seaborn as sns
```

```python
titanic=sns.load_dataset("titanic")
titanic
```

# count plot

```python
sns.countplot(x='pclass',data=titanic)
```

```python
sns.countplot(x='embarked',data=titanic)
```

```python
# for horizontal bar graph
sns.countplot(y='class',data=titanic)
```

```python
#bifergating
sns.countplot(x='class',hue='who',data=titanic)
```

```python
#scatter plot
sns.scatterplot(x='pclass',y='class',data=titanic)
```

```python
sns.scatterplot(x='fare',y='age',data=titanic)
```

```python
# create a horizontal count plot for who and bifergate according to the sex
sns.countplot(y='who',hue='sex',data=titanic)
```

# box plot

```python
sns.boxplot(x='pclass',y='age',data=titanic)
```

```python
sns.boxplot(x='fare',y='embarked',data=titanic)
```

# Reg plot

```python
sns.regplot(x='age',y='fare',data=titanic)
```

# colormap

```python
sns.countplot(x='class',data=titanic,palette=['red','green','blue'])
```

```python
sns.countplot(x='class',data=titanic,palette='Greens')
```

```python
sns.countplot(y='class',data=titanic,palette='Purples')
```

# heatmap

heatmaps plots data as a rectangular colour encoded matrix which is customizable

```python
data=[[100,66,88],
     [77,100,44],
     [33,11,100]]

sns.heatmap(data)
```

```python
data=[[100,66,88],
     [77,100,44],
     [33,11,100]]

sns.heatmap(data, annot=True)
```

```python
# darker the colour,lower is the correlation and
# lighter the colour,higher is the correlation
-in the case of heat map you should have all numerical data
```
