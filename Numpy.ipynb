# Array Inspection



```python
# one dimension
a=np.array([10,20,30,40])
print(a)
print(a.size)# the number of elements that are present in the array
print(a.shape)
```

```python
print(a.ndim) # no of dimensions
```

```python
print(a.shape)
```

```python
#zero dimension
a=np.array(20)
print(a)
print(a.ndim)
print(a.shape)
print(type(a))
```

```python
# two dimension
a=np.array([[10,20,30],
          [20,30,40]])
a
```

```python
print(a.ndim)
print(a.shape)
print(a.size)
print(a.dtype)
```

```python
#three dimension
import numpy as np
a=np.array([
            [[20,30,40],
            [10,20,30],
             [1,2,3]]
])
print(a.ndim)
print(a.size)
print(a.shape)
print(a.dtype)
```

 here 1 represents the no.of array or no.of 2d array or no.of depth
